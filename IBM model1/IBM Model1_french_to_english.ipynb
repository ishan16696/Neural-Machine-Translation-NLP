```python
DEBUG = True
words_to_remove=("a","an","the","of","!")
```

```python
def file_to_read(path1,path2,path3):
    file_train_x=open(path1)
    file_train_y=open(path2)
    output = open(path3)
    return file_train_x,file_train_y,output
    
```

```python
file_train_x,file_train_y,output= file_to_read("Data/french.fe","Data/english.en","Data/alignment_vector")
```

```python
fren_sent=file_train_x.read().split('\n')
eng_sent=file_train_y.read().split('\n')
output_vector= output.read().split('\n')
if DEBUG:
    print(eng_sent[0])
    print(fren_sent[0])
```

```python
english=[]
for sent in eng_sent:
    english.append(sent.lower())
    
french=[]
for sent in fren_sent:
    french.append(sent.lower())
```

```python
length=len(french)
if DEBUG:
    print(french[0])
    print(english[0])
```

```python
data=[]
for i in range(length):
    temp_dict={'eng':english[i],'fr':french[i]}
    data.append(temp_dict)
```

```python
if DEBUG:
    print(data[:3])
```

```python
#unique_eng_temp = []
unique_eng=[]
for i in range(length):
    each_sent_list=english[i].split(' ')
    for word in range(len(each_sent_list)):
        if each_sent_list[word] not in unique_eng:
            #unique_eng_temp.append(each_sent_list[word])
            unique_eng.append(each_sent_list[word])
            
```

```python
unique_fr = []
for i in range(length):
    each_sent_list=french[i].split(' ')
    for word in range(len(each_sent_list)):
        if each_sent_list[word] not in unique_fr:
            unique_fr.append(each_sent_list[word])
```

```python
# unique_eng=[]
# for word in unique_eng_temp:
#     if word not in words_to_remove:
#         unique_eng.append(word)
```

```python
if DEBUG:
    print(len(unique_eng),len(unique_fr))
```

```python
#Initializing t(e|f) uniformly
t={}
for hi in unique_fr:
    for en in unique_eng:
        tup= (en,hi)
        t[tup]=1.0/len(unique_eng)
```

```python
num_iterations = 50
s_total = {}
#temp_array = np.zeros(shape=(len(unique_fr),len(unique_eng)))
if DEBUG:
    for e in data:
        print(e['eng'].split(' '))
        print(e['fr'].split(' '))
        break
```

```python
for loop in range(num_iterations):
    if DEBUG:
        print(loop)
    
    count={}
    total={}
    
    for hi in unique_fr:
        total[hi] = 0.0
    
    for hi in unique_fr:
        for eg in unique_eng:
            tup=(eg,hi)
            count[tup] = 0.0

    for sp in data:
        eng_sent=sp['eng'].split(' ')
        fr_sent=sp['fr'].split(' ')

        
        for eword in eng_sent:
            s_total[eword] = 0.0
#             if eword in words_to_remove:
#                 continue
            for hword in fr_sent:    
                s_total[eword] =s_total[eword]+ t[(eword, hword)]

        
        for eword in eng_sent:
#             if eword in words_to_remove:
#                 continue
            for hword in fr_sent:
                count[(eword, hword)] =count[(eword, hword)]+ t[(eword, hword)]/s_total[eword]
                total[hword] += t[(eword, hword)]/s_total[eword]

    # update probabilities back to t
    for hword in unique_fr:
        for eword in unique_eng:
            t[(eword, hword)] = count[(eword, hword)]/total[hword]
        
```

```python
count=0
meri_total_length=0
given_total_length=0
preicision=0.0
recall=0.0
for i in range(0,len(output_vector)):
    check=output_vector[i].split(" ")
    
    tocheck= myaligned_vector[i]
    meri_total_length+=len(tocheck)
    given_total_length+=len(check)
    
    count+=len(set(tocheck)&set(check))
    

preicision= count/meri_total_length*100
recall=count/given_total_length*100
print(count)
print("Accuracy :",count/len(unique_fr)*100)
print("Precision :",preicision)
print("Recall :",recall)
```
